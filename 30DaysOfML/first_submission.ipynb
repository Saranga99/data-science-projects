```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split

#model and evaluation 
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv("train_folds.csv")
df_test=pd.read_csv("test.csv")
```

```python
df.head()
```

```python
ordinal_encoder=OrdinalEncoder()
```

```python
for fold in range(5):
    X_train=df[df.kfold!=fold].reset_index(drop=True)
    X_test=df[df.kfold==fold].reset_index(drop=True)
    
    #filter object columns by name
    object_columns=[col for col in X_train.columns if "cat" in col]
    
    X_train[object_columns]=ordinal_encoder.fit_transform(X_train[object_columns])
    X_test[object_columns]=ordinal_encoder.fit_transform(X_test[object_columns])
    
```
