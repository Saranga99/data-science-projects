```python
import requests
```

```python
res = requests.get('http://classics.mit.edu/Antoninus/meditations.mb.txt')
```

```python
data = res.text
```

```python
data = data.split('Translated by George Long')[1]
```

```python
import re
```

```python
data = re.sub('-{2,}', '', data)
```

```python
data = re.sub('BOOK [A-Z]+\n', '', data)
```

```python
print(data)
```

```python
data = data.split('THE END')[0]
```

```python
data = data.split('\n\n')
```

```python
data
```

```python
data = [sentence.replace('\n', ' ') for sentence in data]
```

```python
data
```

```python
data = [sentence.strip() for sentence in data]
```

```python
data = [sentence for sentence in data if len(sentence) > 0]
```

```python
data
```

```python
with open('../../data/meditations.txt', 'w') as f:
    for sentence in data:
        f.write(sentence + '\n')
```
